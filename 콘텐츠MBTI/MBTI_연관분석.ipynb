```python
import pandas as pd

df = pd.read_csv('고객별_MBTI.csv')

df_cus_2021 = pd.read_csv('iMBANK_CUSTOMER_DATA_2021(K-DigitalTraining).csv', encoding='cp949')
df_cus_2022 = pd.read_csv('iMBANK_CUSTOMER_DATA_2022(K-DigitalTraining).csv', encoding='cp949')
df_cus_2023 = pd.read_csv('iMBANK_CUSTOMER_DATA_2023(K-DigitalTraining).csv', encoding='cp949')
```

```python
a = pd.concat([df_cus_2021, df_cus_2022])
result = pd.concat([a, df_cus_2023])
result
```

```python
result[(result['기준년월'] == 202301) & (result['고객ID'] == 2455099113)]
```

```python
# df1= df[['거래년월', '고객ID', 'MBTI']]
df.rename(columns={'거래년월' : '기준년월'}, inplace=True)
```

```python
df
```

```python
a = pd.merge(result, df, on = ['기준년월', '고객ID'], how = 'inner')
```

```python
a.isnull().sum()
```

```python
a
```

```python
mbti_count = a.groupby('고객ID')['MBTI'].nunique().reset_index()

# 결과를 보기 좋게 열 이름 변경
mbti_count.columns = ['고객ID', 'MBTI_고유_개수']
mbti_count.sort_values(by = 'MBTI_고유_개수', ascending=False)
mbti_count.describe()
```

```python
a['연령대'].value_counts()
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

plt.rcParams['font.family'] = 'AppleGothic'
import warnings
warnings.filterwarnings('ignore')

b = a.pivot_table(index = '연령대',
              columns='E_I',
              values='대출금액',
              aggfunc='count')
b['총합'] = b.sum(axis=1)

# 각 MBTI 비율(%) 계산
df_percentage = b.div(b['총합'], axis=0) * 100

# '총합' 컬럼 제거 (필요시 유지)
df_percentage = df_percentage.drop(columns=['총합'])
df_percentage = df_percentage.reset_index()

x = df_percentage['연령대']
E = df_percentage['E']
I = df_percentage['I']

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x, E, label='E', color='skyblue')
ax.bar(x, I, label='I', bottom=E, color='lightcoral')

ax.set_ylabel('비율 (%)')
ax.set_xlabel('연령대')
ax.set_title('연령대별 E-I 누적 분포')
ax.legend()

plt.show()
```
